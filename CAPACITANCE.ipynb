## Apex complex capacitance for fixed values of z and h

We define the apex complex capacitance my means of the following equation

$$
C^{*}_{APEX}= 2 \pi \epsilon_{0} R \ln[1+ \frac{R(1-\sin{\theta_{0}})}{z+ \frac{h}{\epsilon^{*}}}]
$$

Let

$$
\epsilon_{0}= 8.85\text{·}10^{-12} \: \text{F·m}^{-1}
$$
$$
R=20 \: \text{nm}
$$
$$
\theta_{0}= 18^{\text{o}}
$$

In which z will be 20 nm and h 150 nm

```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
# We define the Havriliak Negami function parameters
pi=np.pi
tauHN=0.001
deps = 1.0
b = 0.5
c = 1.0
einf = 3.0
freqs = np.logspace(-3,9,140)
omega=2*np.pi*freqs
estar=einf + deps*(1+(1j*omega*tauHN)**b)**-c
```

```python
# We define the Capacitance parameters
e0=8.85E-12
R=20E-9
theta0=18*pi/180
h=150E-9
z=20E-9
```

```python
# We define the Capacitance function
def CAP(R,h,z):
    capac_apex=2*pi*e0*R*np.log(1+ R*(1-np.sin(theta0))/(z+ (h/estar)))
    return capac_apex
```

```python
#Real part of the apex capacitance 
capac_apex = CAP(R,h,z) #Using the parameters defined previously
plt.semilogx(freqs,capac_apex.real/1E-12,'k-')
plt.xlabel("f (Hz)")
plt.ylabel("Capacitance (pF)")
```

```python
#Imaginary part of the apex capacitance 
capac_apex = CAP(R,h,z) #Using the parameters defined previously
plt.semilogx(freqs,-capac_apex.imag/1E-12,'k-')
plt.xlabel("f (Hz)")
plt.ylabel("Capacitance (pF)")
```

```python
#Plot delta phase for the given values
plt.semilogx(freqs,np.arctan(-capac_apex.imag/capac_apex.real)*180/pi)
plt.xlabel("f (Hz)")
plt.ylabel("delta capacitance (deg)")
```

```python
#Plot tan_delta for the given values
plt.semilogx(freqs,(-capac_apex.imag/capac_apex.real))
plt.xlabel("f (Hz)")
plt.ylabel("tan delta")
```

## Cantilever complex capacitance 

We define the cantilever complex capacitance my means of the following equation (please change accordingly)
$$
C^{*}_{cantilever}= (f_{BE}(z+T)·(\frac{\epsilon_{0}·W}{\tan(\alpha)}·\ln(1+ \frac{L·\sin(\alpha)}{z+T}))^{-1}+(\frac{\epsilon_{0}·\epsilon^{*}·A_{cantilever}}{h})^{-1})^{-1}
$$

Let

$$
\epsilon_{0}= 8.85\text{·}10^{-12} \: \text{F·m}^{-1}
$$
$$
W=30 \: \text{micrometers}
$$
$$
\alpha = 15^{\text{o}}
$$
$$
L=125 \: \text{micrometers}
$$
$$
h=150 \: \text{nm}
$$
$$
T=15 \: \text{micrometers}
$$
$$
A_{cantilever}=L·W
$$

From now on we will change the parameter "H" to "T" to avoid confusion with the lowercase "h".

For the border effects function let's start by the simple approximation:
$$
f_{\text{BE}} = 2
$$

```python
# We define the Cantilever parameters
e0=8.85E-12
W=30E-6
alpha=15*pi/180
L=125E-6
h=150E-9
z=20E-9
T=15E-6
A=L*W
f=2
```

```python
# We define the Cantilever function
def CANT(W,L,h,T,z):
    capac_cant=(f*(e0*W*np.log(1+(L*np.sin(alpha)/(z+T)))/np.tan(alpha))**-1+(e0*estar*A/h)**-1)**-1
    return capac_cant
```

```python
#Real part of the cantilever capacitance 
capac_cant = CANT(W,L,h,T,z) #Using the parameters defined previously
plt.semilogx(freqs,capac_cant.real,'k-')
plt.xlabel("f (Hz)")
plt.ylabel("Real part")
```

```python
#Imaginary part of the cantilever capacitance 
capac_cant = CANT(W,L,h,T,z) #Using the parameters defined previously
plt.semilogx(freqs,-capac_cant.imag,'k-')
plt.xlabel("f (Hz)")
plt.ylabel("Imaginary part")
```

```python
#Plot delta phase for the given values
plt.semilogx(freqs,np.arctan(-capac_cant.imag/capac_cant.real)*180/pi)
plt.xlabel("f (Hz)")
plt.ylabel("delta capacitance (deg)")
```

```python
#Plot tan_delta for the given values
plt.semilogx(freqs,(-capac_cant.imag/capac_cant.real))
plt.xlabel("f (Hz)")
plt.ylabel("tan delta")
```
